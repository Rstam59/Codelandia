```python
import numpy as np
import plotly.graph_objects as go

# Generate data
x = np.linspace(-10, 10, 10)
y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# Create a scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=X.flatten(),
    y=Y.flatten(),
    z=Z.flatten(),
    mode='markers'
)])

# Show the plot
fig.show()
```
